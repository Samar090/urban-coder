```python
from matplotlib import pyplot as pl
from matplotlib import dates as da
from datetime import datetime,timedelta

pl.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
pl.plot_date(dates,y,linestyle='solid')

pl.gcf().autofmt_xdate() #gcf: Get current figure , autofmt:Used to format the date on x-axis

date_format=da.DateFormatter('%b,%d %y')

pl.gca().xaxis.set_major_formatter()# gca:Get current axis , then we have specified the axis , set_major_foematter():Used set the formatter to the ticker of the specified axis

pl.tight_layout()

pl.show()
```
